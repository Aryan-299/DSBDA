```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("forestfires.csv")
df
```

```python
plt.hist(df['FFMC'], bins = 10, color = 'yellow', edgecolor = 'black')
plt.xlabel('FFMC')
plt.ylabel('Frequency')
plt.title('FFMC Distribution')
plt.show()
```

```python
plt.plot(df['DMC'], 'o' , markersize = 3 , color='orange')
plt.xlabel('DMC')
plt.ylabel('Frequency')
plt.title('DMC Scatterplot')
plt.show()
```

```python
gender_counts = df['day'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color = ['red','yellow'])
plt.xlabel('Days')
plt.ylabel('Count')
plt.title('Heart Disease Count by Day')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='day', y='DMC', data=df)
plt.title('Cholesterol Levels Over Age')
plt.xlabel('Age')
plt.ylabel('Cholesterol Level (mg/dl)')
plt.show()
```

```python
plt.subplot(2, 2, 1)
sns.boxplot(df['wind'], color='skyblue')
plt.title('Month Box Plot')
```

```python
# Histogram
plt.subplot(2, 2, 2)
plt.hist(df['wind'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
```

```python
plt.subplot(2, 2, 3)
plt.scatter(df['wind'], df['day'], color='orange', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age vs Cholesterol')
plt.tight_layout()
plt.show()
```
